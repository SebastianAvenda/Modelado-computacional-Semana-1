### Ejercicio 1 (resuelto)

Calcule la siguiente sumatoria:

$$
\sum_{k=1}^{n} k
$$

#### Solución matemática

Se utiliza la fórmula de la suma de los primeros \( n \) números naturales:

$$
\sum_{k=1}^{n} k = \frac{n(n+1)}{2}
$$

#### Solución en Python


```python
def suma_naturales(n):
    return n * (n + 1) // 2

print(suma_naturales(10))
```

### Ejercicio 2 (resuelto)

Calcule la siguiente sumatoria:

$$
\sum_{k=1}^{n} (2k - 1)
$$

#### Solución matemática

Esta expresión corresponde a la suma de los primeros \( n \) números impares:

$$
\sum_{k=1}^{n} (2k - 1) = n^2
$$

#### Solución en Python


```python
def suma_segunda(n):
    return sum(k**2 for k in range(1, n+1))

print(suma_segunda(10))
```

## II. Ejercicios de Diccionarios en Python

1. Cree un diccionario con los nombres de cinco estudiantes como claves y sus calificaciones finales como valores. Calcule el promedio.

---



```python
diccionario_estudiantes = {
    "Sebastian": 30,
    "Leon": 35,
    "Diaz": 40,
    "Rojas": 45,
    "Camilo": 50,
}
promedio = sum(diccionario_estudiantes.values()) / len(diccionario_estudiantes)
print(promedio)
```

2. Dado un diccionario de productos y precios, incremente todos los precios en un 10%.
---

```python
diccionario_productos={
    "banano": 3000.0,
    "mango": 5000.0,
    "papaya": 4000.0,
    "limon": 1000.0,
}
for incremento in diccionario_productos:
    diccionario_productos[incremento] *= 1.10
print("Los nuevos precios incrementados son:")
print(diccionario_productos)
```

3. Escriba un programa que determine la clave asociada al valor máximo en un diccionario numérico.
---

```python
diccionario_numerico={
    "a": 17,
    "b": 4,
    "c": 64,
    "d": 30,
}   
clave_max = max(diccionario_numerico, key=diccionario_numerico.get)
print(clave_max)
```

4. Invierta las claves y valores de un diccionario (asuma que los valores son únicos).
---

```python
claves={
    "a": 17,
    "b": 4,
    "c": 64,
    "d": 30,
}
invertido = {valor: clave for clave, valor in claves.items()}
print(invertido)
```

5. Dado un diccionario con palabras y su frecuencia, elimine las palabras con frecuencia menor que 3.
---

```python
frecuencia = {
    "a": 5,
    "b": 7,
    "c": 2,
    "d": 1,
    "e": 2,
}
frecuencia={palabra: frecuencia for palabra, frecuencia in frecuencia.items() if frecuencia >= 3}
print ("El diccionario sin palabras con frecuencia mayores o iguales que 3:", frecuencia)
```

6. Combine dos diccionarios sumando los valores de las claves comunes.
---

```python
diccionario1= {"a": 1, "b": 2, "c": 3}
diccionario2= {"b": 4, "c": 5, "d": 6}
diccionario_suma = {clave: diccionario1.get(clave, 0) + diccionario2.get(clave, 0) for clave in set(diccionario1) | set(diccionario2)}
print("suma de diccionarios", diccionario_suma)
```

---

## III. Ejercicios de Multiplicación de Diccionarios en Python

(Se entiende por multiplicación la operación valor a valor entre diccionarios con claves comunes).

1. Dados dos diccionarios con las mismas claves y valores numéricos, construya un tercer diccionario donde cada valor sea el producto de los valores correspondientes.
---

```python
valor1= {"a": 17, "b": 4, "c": 64, "d": 30}
valor2= {"a": 56, "b": 35, "c": 11, "d": 24}
producto= {k: valor1[k] * valor2[k] for k in valor1}
print("Producto de los valores de los diccionarios:", producto)
```

---
2. Multiplique todos los valores de un diccionario por un escalar dado.
---

```python
valor1= {"a": 17, "b": 4, "c": 64, "d": 30}
escalar= 2
valor2= {k: valor1[k] * escalar for k in valor1}
print("Producto de los valores del diccionario por el escalar:", valor2)
```

3. Dado un diccionario que representa dos vectores con las mismas claves, calcule su producto punto:
   
   $$
   \sum_{i} v_i \cdot w_i
   $$
---

```python
vector1= {"a": 17, "b": 4, "c": 64, "d": 30}
vector2= {"a": 56, "b": 35, "c": 11, "d": 24}
producto_punto = sum(vector1[k] * vector2[k] for k in vector1)
print("Producto punto de los dos vectores:", producto_punto)
```

4. Dado un diccionario de precios y otro de cantidades, calcule el costo total por producto y el costo total general.
---

```python
precios={"pan": 10000, "leche": 25000, "huevos": 11000}
cantidades={"pan": 2, "leche": 5, "huevos": 12}
costos= {k: precios[k] * cantidades[k] for k in precios}
total = sum(costos.values())
print("Costo por producto:", costos)
print("Costo total:", total)
```
